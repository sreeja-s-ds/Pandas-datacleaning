```python
df.AREA.unique()
```

```python
df["AREA"]=df["AREA"].apply(lambda i:"Chrompet" if (i=="Chrompt" or i=="Chrmpet" or i=="Chormpet") else i) # Chrompet spellcheck
```

```python
df["AREA"]=df["AREA"].apply(lambda i:"Anna Nagar" if (i=="Ana Nagar" or i=="Ann Nagar") else i) #Anna Nagar spellcheck
```

```python
df["AREA"] = df["AREA"].replace({
    "Karapakam":"Karapakkam",
    "Adyr":"Adyar",
    "Velchery": "Velachery",
    "KKNagar": "KK Nagar",
    "TNagar" : "T Nagar"
}) # Remaining spellcheck
```

```python
df.AREA.unique() # Spellings correction checked
```

```python
from sklearn.preprocessing import LabelEncoder

a = LabelEncoder()
df["AREA_encoded"] = a.fit_transform(df["AREA"])
```

```python
df # Column Area encoded & added to the last
```

```python
df["DATE_SALE"] = pd.to_datetime(df["DATE_SALE"], dayfirst=True, errors="coerce")
```

```python
df["DATE_SALE_DETAILS"] = (
    df["DATE_SALE"].dt.strftime('%d-%m-%Y') +
    " | Day:" + df["DATE_SALE"].dt.day.astype(str) +
    " | Month:" + df["DATE_SALE"].dt.month.astype(str) +
    " | Year:" + df["DATE_SALE"].dt.year.astype(str) +
    " | Week:" + df["DATE_SALE"].dt.isocalendar().week.astype(str)
) # This line constructs a new column like:"21-02-2020 | Day:21 | Month:2 | Year:2020 | Week:8"
```

```python
df
```

```python
df["DATE_BUILD"] = pd.to_datetime(df["DATE_BUILD"], dayfirst=True, errors="coerce")

df["DATE_BUILD_DETAILS"] = (
    df["DATE_BUILD"].dt.strftime('%d-%m-%Y') +
    " | Day:" + df["DATE_BUILD"].dt.day.astype(str) +
    " | Month:" + df["DATE_BUILD"].dt.month.astype(str) +
    " | Year:" + df["DATE_BUILD"].dt.year.astype(str) +
    " | Week:" + df["DATE_BUILD"].dt.isocalendar().week.astype(str)
) # DATE_BUILD_DETAILS column created with required conversions
```

```python
df
```

```python
df["BUILDING_AGE"] = (
    (df["DATE_SALE"] - df["DATE_BUILD"]).dt.days // 365
).astype(str) + " years | " + (
    (df["DATE_SALE"] - df["DATE_BUILD"]).dt.days
).astype(str) + " days"
# Building age calculated
```
